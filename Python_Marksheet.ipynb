```python
# Student Name: Cheah Wan Xin
# Student ID  : B2101241
# Subject     : BDA100 Introduction to Programming 
# Assessment  : Assignment 2 (10%)

class PythonMarksheet: 
    "fields: student_ID, student_name, test1, assignment1, test2, assignment2, final_exam"
    
    def __init__(self, student_ID, student_name, test1, assignment1, test2, assignment2, final_exam): #Constructors
        self.student_ID = student_ID
        self.student_name = student_name
        self.test1 = test1 
        self.assignment1 = assignment1
        self.test2 = test2 
        self.assignment2 = assignment2
        self.final_exam = final_exam
    
#************************************************************************************************* GETTER METHOD ****
    def getStudent_ID(self): 
        return self.student_ID
    
    def getStudent_name(self): 
        return self.student_name
    
    def getTest1(self): 
        return self.test1
    
    def getAssignment1(self): 
        return self.assignment1
    
    def getTest2(self): 
        return self.test2
    
    def getAssignment2(self): 
        return self.assignment2
    
    def getFinal_exam(self): 
        return self.final_exam 
    
#************************************************************************************************* SETTER METHOD ****

    def setStudent_ID(self, student_ID): 
        self.student_ID = student_ID
        
    def setStudent_name(self, student_name): 
        self.student_name = student_name
    
    def setTest1(self, test1): 
        self.test1 = test1
        
    def setAssignment1(self, assignment1): 
        self.assignment1 = assignment1
        
    def setTest2(self, test2): 
        self.test2 = test2
    
    def setAssignment2(self, assignment2): 
        self.assignment2 = assignment2
        
    def setFinal_exam(self, final_exam): 
        self.final_exam = final_exam 
        
#************************************************************************** Total Mark & Final Grade Calculation ****
    def total_mark(self): 
        total_mark = ((self.getTest1() + self.getAssignment1() + self.getTest2() + self.getAssignment2())*0.1) + ((self.getFinal_exam())*0.6) 
        return total_mark 
    
    def final_grade(self): 
        if self.total_mark() >= 85: 
            return 'HD1'
        elif self.total_mark() < 85 and self.total_mark() >= 80:
            return 'HD2'
        elif self.total_mark() < 80 and self.total_mark() >= 75: 
            return 'DI1'
        elif self.total_mark() < 75 and self.total_mark() >= 70: 
            return 'DI2'
        elif self.total_mark() < 70 and self.total_mark() >= 65:
            return 'CR1'
        elif self.total_mark() < 65 and self.total_mark() >= 60:
            return 'CR2'
        elif self.total_mark() < 60 and self.total_mark() >= 55:
            return 'PS1'
        elif self.total_mark() < 55 and self.total_mark() >= 50:
            return 'PS2'
        elif self.total_mark() < 50:
            return 'F'
        
#*********************************************************************************************** Strings Method *****       
    def __str__(self): 
        return "\nStudent ID  : {} \n"\
                 "Student Name: {} \n\n" \
                 "Student Assessment Marks: \n" \
                 "Test 1      : {} \n" \
                 "Assignment 1: {} \n" \
                 "Test 2      : {} \n" \
                 "Assignment 2: {} \n" \
                 "Final Exam  : {} \n\n" \
                 "Total Marks : {:.2f}% \n" \
                 "Final Grade : {} \n".format(self.getStudent_ID(), self.getStudent_name(), self.getTest1(), \
                                              self.getAssignment1(), self.getTest2(), self.getAssignment2(), \
                                              self.getFinal_exam(), self.total_mark(), self.final_grade())
```
