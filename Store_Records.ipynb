```python
# Student Name: Cheah Wan Xin
# Student ID  : B2101241
# Subject     : BDA100 Introduction to Programming 
# Assessment  : Assignment 2 (10%)

from ipynb.fs.full.Python_Marksheet import PythonMarksheet

#*************************************************************************************** Store & Update Records *****

class StoreRecords: 
    
    def __init__(self):
        self.__index = 0 
        self.__marksheet = {}

    def add_student_records(self, marksheet):
        self.__marksheet[self.__index] = marksheet
        self.__index += 1
        
    def num_of_records(self): 
        return len(self.__marksheet)
    
    def verifyStudent_ID(self, student_id): 
        for PythonMarksheet in self.__marksheet.values(): 
            if PythonMarksheet.getStudent_ID() == student_id: 
                return True
                break
        return False

#********************************************************************************* Search, Edit, Delete Records *****
    def search_student_records(self, student_id):
        count = 0
        for PythonMarksheet in self.__marksheet.values(): 
            if PythonMarksheet.getStudent_ID() == student_id: 
                record2 = self.__marksheet[count]
                return record2
                break
            count += 1
    
    def edit_student_name(self, student_id, new_Name): 
        for PythonMarksheet in self.__marksheet.values():
            if PythonMarksheet.getStudent_ID() == student_id: 
                PythonMarksheet.setStudent_name(new_Name)
                break 
    
    def edit_student_records(self, student_id, new_test1, new_assignment1, new_test2, new_assignment2, new_final_exam):
        for PythonMarksheet in self.__marksheet.values(): 
            if PythonMarksheet.getStudent_ID() == student_id: 
                PythonMarksheet.setTest1(new_test1)
                PythonMarksheet.setAssignment1(new_assignment1)
                PythonMarksheet.setTest2(new_test2)
                PythonMarksheet.setAssignment2(new_assignment2)
                PythonMarksheet.setFinal_exam(new_final_exam)
                break
                
    def remove_student_records(self, student_id): 
        count1 = 0
        for PythonMarksheet in self.__marksheet.values(): 
            if PythonMarksheet.getStudent_ID() == student_id: 
                del self.__marksheet[count1]
                break
            count1 += 1

#******************************************************************************** Display, Load, Export Records *****
    def display_all_records(self): 
        records = ''
        for PythonMarksheet in self.__marksheet.values(): 
            records = records + str(PythonMarksheet) + '\n'
        return records

    def verifyFile_name(self, year, semester): 
        try: 
            open(f'PythonMarksheet_{year}_Semester{semester}.txt', 'r')
            return True
        except: 
            return f"Student's marks record in Semester {semester} ({year}) not found!"
    
    def load_records(self, year, semester): 
        loadFile = open(f'PythonMarksheet_{year}_Semester{semester}.txt', 'r')

        for sentence in loadFile:
            text = sentence.split("--")

            student_id = int(text[1])
            name = text[2]
            test_1 = float(text[3])
            assignment_1 = float(text[4])
            test_2 = float(text[5])
            assignment_2 = float(text[6])
            Final_Exam = float(text[7])

            print("\nStudent ID  : {} \n"\
                  "Student Name: {} \n\n" \
                  "Student Assessment Marks: \n" \
                  "Test 1      : {} \n" \
                  "Assignment 1: {} \n" \
                  "Test 2      : {} \n" \
                  "Assignment 2: {} \n" \
                  "Final Exam  : {} ".format(student_id, name, test_1, assignment_1, test_2, assignment_2, Final_Exam))

            new_student_records = PythonMarksheet(student_id, name, test_1, assignment_1, test_2, assignment_2, Final_Exam)
            self.add_student_records(new_student_records)

            total_mark = '%.2f'%new_student_records.total_mark()
            final_grade = new_student_records.final_grade()

            print(f"\nTotal Mark : {total_mark}%")
            print("Final Grade: ", final_grade)
        
        loadFile.close()
    
    def save_records(self, year, semester): 
        count2 = 0
        saveFile = open(f'PythonMarksheet_{year}_Semester{semester}.txt', 'w')
        for x in self.__marksheet.values(): 
            count2 += 1
            saveFile.write("{}--{}--{}--{}--{}--{}--{}--{}\n".format(count2, x.getStudent_ID(), x.getStudent_name(), x.getTest1(), \
                                                                     x.getAssignment1(), x.getTest2(), x.getAssignment2(), \
                                                                     x.getFinal_exam()))
            
        saveFile.close()

#************************************************************************************* Grade Distribution Table *****
    def grade_distribution(self):
        countF = countPS2 = countPS1 = countCR2 = countCR1 = countDI2 = countDI1 = countHD2 = countHD1 = 0
        for PythonMarksheet in self.__marksheet.values(): 
            if PythonMarksheet.final_grade() == 'F':
                countF += 1
            elif PythonMarksheet.final_grade() == 'PS2':
                countPS1 += 1
            elif PythonMarksheet.final_grade() == 'PS1':
                countPS2 += 1
            elif PythonMarksheet.final_grade() == 'CR2':
                countCR2 += 1
            elif PythonMarksheet.final_grade() == 'CR1':
                countCR1 += 1
            elif PythonMarksheet.final_grade() == "DI2":
                countDI2 += 1
            elif PythonMarksheet.final_grade() == 'DI1':
                countDI1 += 1
            elif PythonMarksheet.final_grade() == 'HD2':
                countHD2 += 1
            elif PythonMarksheet.final_grade() == 'HD1':
                countHD1 += 1
        
        print("\nNumber of students with \n" \
                f"Grade F   : {countF}\n" \
                f"Grade PS2 : {countPS2}\n" \
                f"Grade PS1 : {countPS1}\n" \
                f"Grade CR2 : {countCR2}\n" \
                f"Grade CR1 : {countCR1}\n" \
                f"Grade DI2 : {countDI2}\n" \
                f"Grade DI1 : {countDI1}\n" \
                f"Grade HD2 : {countHD2}\n" \
                f"Grade HD1 : {countHD1}\n")
```
