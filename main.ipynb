```python
# Student Name: Cheah Wan Xin
# Student ID  : B2101241
# Subject     : BDA100 Introduction to Programming 
# Assessment  : Assignment 2 (10%)

from ipynb.fs.full.Python_Marksheet import PythonMarksheet
from ipynb.fs.full.Store_Records import StoreRecords

#************************************************************************************************** Menu Option *****

def menu_option():
    while True: 
        try: 
            print('_'*66)
            print("Enter 0 to exit \n" \
                  "Enter 1 to add student's marks \n" \
                  "Enter 2 to search student's marks record \n" \
                  "Enter 3 to edit or delete student's marks record \n" \
                  "Enter 4 to display all student's marks record \n" \
                  "Enter 5 to load student's marks record from external file \n" \
                  "Enter 6 to save all students' marks into external file \n" \
                  "Enter 7 to generate grade distribution table")
            print('_'*66)
            choice1 = int(input("Enter your choice: ")) 
            return choice1
            break
        
        except ValueError: 
            print("Invalid input. Please re-enter. Thank You!")
            continue

#*********************************************************************************************** Add New Record *****
record = StoreRecords()

def add_student_marks():
    while True: 
        try: 
            student_ID = int(input("\nEnter student ID  : "))
            student_name = str(input("Enter student name: "))
            
            if len(student_name) < 10: 
                print("Please enter a name with at least 10 characters!")
            else: 
                if student_name.isdigit() == True: 
                    print("Invalid input. Please enter characters!")
                else: 
                    capitalize_name = student_name.title()
                    print(f"\nEnter assessment marks for {capitalize_name}({student_ID}) in 100% format: ")
                    test1 = float(input("Test 1      : "))
                    assignment1 = float(input("Assignment 1: "))
                    test2 = float(input("Test 2      : "))
                    assignment2 = float(input("Assignment 2: "))
                    final_exam = float(input("Final Exam  : "))

                    marksheet = PythonMarksheet(student_ID, capitalize_name, test1, assignment1, test2, assignment2, final_exam)
                    record.add_student_records(marksheet)

                    total_mark = '%.2f'%marksheet.total_mark()
                    final_grade = marksheet.final_grade()

                    print(f"\nTotal Mark : {total_mark}%")
                    print("Final Grade: ", final_grade)
                    break 

        except ValueError: 
            print("\nInvalid input. Please re-enter. Thank You!")
            continue

#*********************************************************************************************** Search Records *****
        
def search_student_marks():
    while True: 
        try: 
            student_id = int(input("\nEnter student ID to search: "))
            if record.verifyStudent_ID(student_id) == True: 
                print(record.search_student_records(student_id))
            else: 
                print("\nStudent ID not found!")
            break 
        except ValueError: 
            print("\nInvalid input. Please re-enter. Thank You!")
            continue 

#**************************************************************************************** Edit & Delete Records *****

def edit_student_marks():
    while True:
        try: 
            student_id = int(input("\nEnter student ID to edit: "))
            if record.verifyStudent_ID(student_id) == True: 
                print("\nEnter 1 to edit student's name \n" \
                      "Enter 2 to edit student's marks \n" \
                      "Enter 3 to delete student's marks \n")
                choice2 = int(input("Enter your choice: "))

                if choice2 == 1: 
                    new_Name = str(input("\nEnter student's new name: "))
                    if len(new_Name) < 10: 
                        print("Please enter a name with at least 10 characters!")
                    else: 
                        if new_Name.isdigit() == True: 
                            print("Invalid input. Please enter characters!")
                        else: 
                            record.edit_student_name(student_id, new_Name.title())
                            print("\nPlease choose menu option 6 to successfully update student's name!")
                            break 

                elif choice2 == 2: 
                    print(f"\nEnter new assessment marks for {student_id} in 100% format: ")
                    new_test1 = float(input("Test 1      : "))
                    new_assignment1 = float(input("Assignment 1: "))
                    new_test2 = float(input("Test 2      : "))
                    new_assignment2 = float(input("Assignment 2: "))
                    new_final_exam = float(input("Final Exam  : "))
                    record.edit_student_records(student_id, new_test1, new_assignment1, new_test2, new_assignment2, new_final_exam)
                    print("\nPlease select menu option 6 to successfully update student's marks!")
                    break 

                elif choice2 == 3: 
                    confirm = str(input(f"\nAre you sure you want to permanently delete {student_id}? (Y/N): "))
                    if confirm == 'Y' or confirm == 'y': 
                        record.remove_student_records(student_id)
                        print("\nPlease choose menu option 6 to successfully delete student's records!")
                        break 
                    elif confirm == 'N' or confirm == 'n': 
                        print("You have chosen not to delete the records.")
                        break
                    else: 
                        print("\nInvalid input. Please re-enter. Thank You!")
                        continue 
            else: 
                print("\nStudent ID not found!")
                break
                
        except ValueError: 
            print("\nInvalid input. Please re-enter. Thank You!")
            continue 
                
#********************************************************************************************** Display Records *****

def display_student_marks():
    if record.num_of_records() == 0: 
        print("\nPlease add a student's marks record!")
    else: 
        print("\nStudent's marks record are displayed as follows: ")
        print(record.display_all_records())
    
#************************************************************************************************* Load Records *****

def load_student_marks():
    while True: 
        try:  
            print("\nEnter a year and semester to load data from: ")
            year = int(input("Year    : "))
            semester = int(input("Semester: "))
            if record.verifyFile_name(year, semester) == True: 
                print('_'*66)
                print("Student's marks record are displayed as follows: ") 
                record.load_records(year, semester)
                print("\nSuccessfully Loaded!")
            else: 
                print(record.verifyFile_name(year, semester))
            break 
            
        except ValueError: 
            print("\nInvalid input. Please re-enter. Thank You!")
            continue 

#************************************************************************************************ Export Records ****

def save_student_marks():
    while True: 
        try: 
            print("\nEnter a year and semester to export data: ")
            year = int(input("Year    : "))
            semester = int(input("Semester: "))
            record.save_records(year, semester)
            print("\nSuccessfully Exported!")
            break 
            
        except ValueError: 
            print("\nInvalid input. Please re-enter. Thank You!")
            continue 

#************************************************************************************* Grade Distribution Table *****

def grade_distribution_table(): 
    if record.num_of_records() == 0: 
        print("\nPlease add a student's marks record!")
    else: 
        record.grade_distribution()

#************************************************************************************************ Main Function *****

def main():  
    print("Welcome to Student's Marksheet Management System!")
    response = menu_option() 
    while response != 0:
        if response == 1: 
            add_student_marks()
        elif response == 2: 
            search_student_marks()
        elif response == 3:
            edit_student_marks()
        elif response == 4:
            display_student_marks()
        elif response == 5:
            load_student_marks()
        elif response == 6: 
            save_student_marks()
        elif response == 7: 
            grade_distribution_table()
        elif response < 0 or response > 7: 
            print("Invalid Input. Please re-enter. Thank You!")
            
        response = menu_option() 
             
    print("You have chosen to exit. Thank You!")
        
main()
```
